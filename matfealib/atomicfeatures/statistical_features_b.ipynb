```python
def statistical_feature_values(compound,feature_collection,feature_name,statistical_functions='all'):
    '''
    this funtion
    '''
    column_names=[]
    atoms = building_blocks(compound,'element')
    # for idx in range(len(atoms)):
    #     column_names.append('{}_{}_{}'.format(feature_name,idx+1,atoms[idx]))
    mat_fea_vals=[]
    for atom in atoms:
        tmp_val = feature_collection.loc[feature_collection['chemical_symbol'] == atom][feature_name].values
        mat_fea_vals.append(tmp_val[0])
    mat_fea_vals_array = np.array(mat_fea_vals)
    
    min_value=mat_fea_vals_array.min()
    max_value=mat_fea_vals_array.max()
    mean_value=mat_fea_vals_array.mean()
    std_value=mat_fea_vals_array.std()
    sum_value=mat_fea_vals_array.sum()
    diff_value=min_value=max_value-min_value
    tmp_dict={'min':min_value, 'max':max_value, 'mean':mean_value, 'std':std_value, 'diff':diff_value, 'sum':sum_value}


    if (statistical_functions=='all'):
        for i in tmp_dict.keys():
            column_names.append(feature_name+'_{}'.format(i))
            # print(column_names)
        return column_names, list(tmp_dict.values())
    elif isinstance(statistical_functions, list):
        tmp_lst=[]
        for i in statistical_functions:
            # print(i)
            column_names.append(feature_name+'_{}'.format(i))
            # print(column_names)
            tmp_lst.append(tmp_dict[i])
        return column_names,tmp_lst
    elif statistical_functions != 'all' and isinstance(statistical_functions, str):
        if statistical_functions not in tmp_dict.keys():
            raise ValueError("The `statistical_functions` parameter should be one of the `min`, `max`, `sum`, `diff`, `mean`, `median` or `all` valuse")
        else:
            tmp_lst=[]
            column_names.append(feature_name+'_{}'.format(statistical_functions))
            tmp_lst.append(tmp_dict[statistical_functions])
            return column_names,tmp_lst

```

```python
def statistical_feature_values_df(dataframe_name,column_name,feature_collection,feature_name,statistical_functions):
    
    dataframe_name[['stat_name','stat_value']]=dataframe_name[column_name].apply(statistical_feature_values,feature_collection=feature_collection
                                                                                              ,feature_name=feature_name,statistical_functions=statistical_functions).tolist()

    column_names=dataframe_name['stat_name'].tolist()[0]
    new_data=pd.DataFrame(dataframe_name['stat_value'].tolist(),columns=column_names,index=dataframe_name.index)
    dataframe_name=pd.concat([dataframe_name,new_data],axis=1)

    
    return dataframe_name.drop(['stat_name','stat_value'],axis=1)
```

```python
def statistical_features(compound,feature_collection,feature='all',statistical_functions='all'):

    if (feature=='all'):
        tmp_all_list=[]
        column_list=feature_collection.keys().drop('chemical_symbol')
        for i in range(len(column_list)):
            tmp_all_list.append(statistical_feature_values(compound,feature_collection,feature_name=(column_list)[i],statistical_functions=statistical_functions))
        return tmp_all_list
    elif isinstance(feature, list):
        tmp_list=[]
        for i in range(len(feature)):
            tmp_list.append(statistical_feature_values(compound,feature_collection,feature_name=feature[i],statistical_functions=statistical_functions))
        return tmp_list
    elif feature != 'all' and isinstance(feature, str):
        return statistical_feature_values(compound,feature_collection,feature_name=feature,statistical_functions=statistical_functions)
    #     return type(arg)
    # if isinstance(arg, pymatgen.core.composition.Composition):
    #     return type(arg)
    # else:
        # return "Not a dataframe"
```

```python
def statistical_features_df(dataframe_name,column_name,feature_collection,feature='all',statistical_functions='all'):
    if (feature=='all'):
        # for i in range(len(feature_collection.keys())):
        tmp_all_list=[]
        column_list=feature_collection.keys().drop('chemical_symbol')
        # print(column_list)
        # tmp_all_df=dataframe_name.copy()
        for i in range(len(column_list)):
            # print(column_list[i])
            dataframe_name=statistical_feature_values_df(dataframe_name,column_name,feature_collection,feature_name=column_list[i],statistical_functions=statistical_functions).copy()
            # print(i)
            # tmp_all_list.append(tmp_all_df)
            # tmp_all_list=pd.concat(tmp_all_list, ignore_index=True)
        # print('ok')
        # return tmp_all_df
        return dataframe_name
    elif isinstance(feature, list):
        tmp_list=[]
        for i in range(len(feature)):
            tmp_list_df=statistical_feature_values_df(dataframe_name,column_name,feature_collection,feature_name=feature[i],statistical_functions=statistical_functions)
        return tmp_list_df
    elif feature != 'all' and isinstance(feature, str):
        return statistical_feature_values_df(dataframe_name,column_name,feature_collection,feature_name=feature,statistical_functions=statistical_functions)
```

```python
def fetch_features(compounds,collection,features='all',column_name=None,statistical_functions='all'):
    if isinstance(compounds, str):
        # return get_features(compounds,collection,features)
        if features !='all' and isinstance(features, str):
            values= [statistical_features(compounds,collection,features ,statistical_functions=statistical_functions)[1]]
            names = statistical_features(compounds,collection,features ,statistical_functions=statistical_functions)[0]
            return pd.DataFrame(values,columns=names,index=[compounds])
        else :
            n_features=len(statistical_features(compounds,collection,features,statistical_functions))
            n_elements=len(statistical_features(compounds,collection,features,statistical_functions)[0][0])
            tmp_array=np.array(statistical_features(compounds,collection,features,statistical_functions))
            names=list(tmp_array[:,0].reshape(1,n_features*n_elements)[0])
            values=tmp_array[:,1].reshape(1,n_features*n_elements)
            return pd.DataFrame(values,columns=names,index=[compounds])
    if isinstance(compounds, list):
        tmp_df=pd.DataFrame(compounds,columns=["formula"])
        # return tmp_df
        return statistical_features_df(tmp_df,"formula",collection,features, statistical_functions=statistical_functions)
    if isinstance(compounds, pd.DataFrame):
        tmp_dataframe_name=compounds.copy(deep=True)
        if (column_name==None):
            raise ValueError("The input is `pd.DataFrame` and `column_name` argument is not given. Please inter materials column name.")
        else :
            return statistical_features_df(tmp_dataframe_name,column_name,collection,features, statistical_functions=statistical_functions)
```

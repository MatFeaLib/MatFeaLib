```python
def statistical_feature_values(compound,feature_collection,feature_name,statistical_name='all'):
    '''
    this funtion
    '''
    column_names=[]
    atoms = building_blocks(compound,'element')
    # for idx in range(len(atoms)):
    #     column_names.append('{}_{}_{}'.format(feature_name,idx+1,atoms[idx]))
    mat_fea_vals=[]
    for atom in atoms:
        tmp_val = feature_collection.loc[feature_collection['chemical_symbol'] == atom][feature_name].values
        mat_fea_vals.append(tmp_val[0])
    mat_fea_vals_array = np.array(mat_fea_vals)
    
    min_value=mat_fea_vals_array.min()
    max_value=mat_fea_vals_array.max()
    mean_value=mat_fea_vals_array.mean()
    std_value=mat_fea_vals_array.std()
    sum_value=mat_fea_vals_array.sum()
    diff_value=min_value=max_value-min_value
    tmp_dict={'min':min_value, 'max':max_value, 'mean':mean_value, 'std':std_value, 'diff':diff_value, 'sum':sum_value}


    if (statistical_name=='all'):
        for i in tmp_dict.keys():
            column_names.append(feature_name+'_{}'.format(i))
            # print(column_names)
        return column_names, list(tmp_dict.values())
    elif isinstance(statistical_name, list):
        tmp_lst=[]
        for i in statistical_name:
            # print(i)
            tmp_lst.append(tmp_dict[i])
        return tmp_lst
    elif statistical_name != 'all' and isinstance(statistical_name, str):
        tmp_lst=[]
        tmp_lst.append(tmp_dict[statistical_name])
        return tmp_lst

```

```python
def statistical_features_df(dataframe_name,column_name,feature_collection,feature='all',statistical_name='all'):
    if (feature=='all'):
        # for i in range(len(feature_collection.keys())):
        tmp_all_list=[]
        column_list=feature_collection.keys().drop('chemical_symbol')
        for i in range(len(column_list)):
            tmp_all_df=statistical_feature_values_df(dataframe_name,column_name,feature_collection,feature_name=(column_list)[i])
        return tmp_all_df
    elif isinstance(feature, list):
        tmp_list=[]
        for i in range(len(feature)):
            tmp_list_df=statistical_feature_values_df(dataframe_name,column_name,feature_collection,feature_name=feature[i])
        return tmp_list_df
    elif feature != 'all' and isinstance(feature, str):
        return statistical_feature_values_df(dataframe_name,column_name,feature_collection,feature_name=feature,statistical_name='all')
```

```python
def fetch_features(compounds, collection, features='all', column_name=None, statistical_name='all'):
    if isinstance(compounds, str):
        # return get_features(compounds,collection,features)
        if features !='all' and isinstance(features, str):
            values= [statistical_features(compounds,collection,features ,statistical_name='all')[1]]
            names = statistical_features(compounds,collection,features ,statistical_name='all')[0]
            return pd.DataFrame(values,columns=names,index=[compounds])
        else :
            n_features=len(statistical_features(compounds,collection,features))
            n_elements=len(statistical_features(compounds,collection,features)[0][0])
            tmp_array=np.array(statistical_features(compounds,collection,features))
            names=list(tmp_array[:,0].reshape(1,n_features*n_elements)[0])
            values=tmp_array[:,1].reshape(1,n_features*n_elements)
            return pd.DataFrame(values,columns=names,index=[compounds])
    if isinstance(compounds, list):
        tmp_df=pd.DataFrame(compounds,columns=["formula"])
        # return tmp_df
        return statistical_features_df(tmp_df,"formula",collection,features, statistical_name='all')
    if isinstance(compounds, pd.DataFrame):
        tmp_dataframe_name=compounds.copy(deep=True)
        if (column_name==None):
            raise ValueError("The input is `pd.DataFrame` and `column_name` argument is not given. Please inter materials column name.")
        else :
            return statistical_features_df(tmp_dataframe_name,column_name,collection,features, statistical_name='all')
```
